```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the sales_data.csv file created earlier
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
# Check basic info
df.info()

# Handle missing values if any
df = df.fillna(0)

# Convert date column if needed
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Basic statistics
print(df.describe())

# Total revenue
print("Total Revenue:", df['Revenue'].sum())

# Best selling product
best = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
print("Best Product:\n", best)

# Average units sold per product
print(df.groupby('Product')['Units_Sold'].mean())
```

```python
# Daily revenue trend
plt.figure(figsize=(8,4))
sns.lineplot(data=df, x='Date', y='Revenue', marker='o')
plt.title('Daily Revenue Trend')
plt.show()

# Product-wise revenue
plt.figure(figsize=(6,4))
sns.barplot(x='Product', y='Revenue', data=df, estimator=sum)
plt.title('Revenue by Product')
plt.show()

# Correlation heatmap
plt.figure(figsize=(6,4)) # Added a figure size for consistency
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm') # Select only numeric columns
plt.title('Correlation Heatmap')
plt.show()
```

```python
### Insights Summary
- Product B generates the highest revenue.
- Sales peak on certain days (visible in line chart).
- Positive correlation between Units_Sold and Revenue.
- Average price remains stable across days.
```
